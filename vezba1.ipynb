Prvi zadatak: Učenici odeljenja IV1 su: Marko, Ana, Jovan, Milica, Sara, Petar i njihove ocene su redom: 4, 5, 3, 5, 2, 4. Predstaviti ocene svakog pojedinačnog učenika stubičastim grafikonom.

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Podaci (ocene od 1 do 5)
data = {
    "Ime": ["Marko", "Ana", "Jovan", "Milica", "Sara", "Petar"],
    "Ocena": [4, 5, 3, 5, 2, 4]
}

df = pd.DataFrame(data)

# 2. Crtanje stubova grafikona
plt.bar(df["Ime"], df["Ocena"], color='orange')
plt.title("Ocene učenika u odeljenju (1-5)")
plt.xlabel("Učenik")
plt.ylabel("Ocena")
plt.ylim(0, 5)
plt.show()
```

Drugi zadatak: Neka Kina u 2025. godini ima 1 412 000 000 stanovnika, a Indija 1 360 000 000. Godišnja stopa rasta stanovništva za Kinu iznosi 0,2%, a za Indiju 0,8%. 

1.Odrediti nakon koliko godina će broj stanovnika Indije prestići broj stanovnika Kine.
2.Grafički prikazati zavisnost broja stanovnika od godina za obe zemlje.

```python
import matplotlib.pyplot as plt

# Početni podaci (u milijardama)
kina = 1.412
indija = 1.360

# Stope rasta (procenti -> decimalni oblik)
rast_kina = 0.002
rast_indija = 0.008

# Početna godina
pocetna_godina = 2025
godina = pocetna_godina

lista_kina = [kina]
lista_indija = [indija]
lista_godina = [godina]

# Simulacija rasta dok Indija ne prestigne Kinu
while indija <= kina:
    godina += 1
    kina = kina * (1 + rast_kina)
    indija = indija * (1 + rast_indija)
    lista_kina.append(kina)
    lista_indija.append(indija)
    lista_godina.append(godina)

# Koliko godina je potrebno
broj_godina = godina - pocetna_godina

print(f"Indija će prestići Kinu za {broj_godina} godina, oko {godina}. godine")

# Grafik – linije rasta
plt.plot(lista_godina, lista_kina, label="Kina", color='red')
plt.plot(lista_godina, lista_indija, label="Indija", color='green')

# Vertikalna linija na tački preseka
plt.axvline(x=godina, color='blue', linestyle='--', label=f"Presek nakon {broj_godina} godina")

plt.title("Rast stanovništva Kine i Indije")
plt.xlabel("Godina")
plt.ylabel("Broj stanovnika (milijarde)")
plt.legend()
plt.grid(True)
plt.xticks(lista_godina)
plt.show()
```
